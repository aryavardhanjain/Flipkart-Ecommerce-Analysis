### Load Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/flipkart_data.csv")
```

### Basic Dataset Structure

```python
print("Shape:", df.shape)
print("\nColumn Names:\n", df.columns.to_list())
df.tail(10)
```

### Data Types and Missing Values

```python
df.info()

df.isna().sum()
```

### Summary statistics

```python
df.describe()
```

### Data Cleaning

```python
df = df.dropna(subset=["retail_price", "discounted_price"])
```

```python
df['product_rating'] = df['product_rating'].replace('No rating available', np.nan)
df['overall_rating'] = df['overall_rating'].replace('No rating available', np.nan)

df['product_rating'] = df['product_rating'].astype(float)
df['overall_rating'] = df['overall_rating'].astype(float)
```

```python
df = df.dropna(subset=['image', 'description'])
```

```python
df['brand'] = df['brand'].fillna('Unknown')
```

### Check for Duplicates

```python
df.duplicated().sum()
```

```python
df['product_name'].value_counts().head(10)
```

```python
# Negative or zero prices
df[df['retail_price'] <= 0]
df[df['discounted_price'] <= 0]

# Discounted price greater than retail
df[df['discounted_price'] > df['retail_price']]
```

```python
df.to_csv("../data/flipkart_cleaned.csv", index=False)
```
